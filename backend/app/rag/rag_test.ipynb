```python
from app.rag.tests.test_suite import TESTS
from app.rag.pipeline import RAGConfig, run_rag
import json

async def run_frontend_sample(limit=None, top_k=4, enable_consult=False):
    out = []
    tests = TESTS if limit is None else TESTS[:limit]
    for t in tests:
        res = await run_rag(
            t['query'],
            config=RAGConfig(top_k=top_k, enable_consult_search=enable_consult),
        )
        out.append({
            'id': t['id'],
            'query': t['query'],
            'guidanceScript': res.get('guidanceScript') or (res.get('guide_script') or {}).get('message') or '',
            'currentSituation': res.get('currentSituation') or [],
            'nextStep': res.get('nextStep') or [],
        })
    print(json.dumps(out, ensure_ascii=False, indent=2))
    return out

await run_frontend_sample()
```
